The data in the file says either 'YES' or 'NO' to signify Oscar winner or non-oscar winner. However, on this visualization it may be confusing to a viewer. Let's try to fix that...

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a box plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='Oscar', y='IMDb Rating', data=df, palette="Set3")
plt.title('Distribution of IMDb Ratings for Oscar-winning vs. Non-Oscar-winning Movies')
plt.xlabel('Oscar Winners')
plt.ylabel('IMDb Rating')
plt.xticks([0, 1], ['Non-Oscar Winners', 'Oscar Winners'])  # Rename x-axis labels
plt.grid(axis='y', linestyle='--', alpha=0.7)
sns.despine(trim=True, left=True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Separating Oscar Winners and Non Oscar Winners in violin plots
plt.figure(figsize=(12, 8))
sns.violinplot(x='Oscar', y='IMDb Rating', data=df, palette="Set3", inner="quartile")
plt.title('Distribution of IMDb Ratings for Oscar-winning vs. Non-Oscar-winning Movies')
plt.xlabel('Oscar Winners')
plt.ylabel('IMDb Rating')
plt.xticks([0, 1], ['Non-Oscar Winners', 'Oscar Winners'])  # Rename x-axis labels
plt.grid(axis='y', linestyle='--', alpha=0.7)
sns.despine(trim=True, left=True)

plt.show()
```
